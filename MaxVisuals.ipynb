```python
import pandas as pd
import numpy as np
```

```python
DATA=pd.read_csv('Log.csv', names=['DateTime','RomanDate','Bookie','Tournement','Team1','Team2','OddsT1','OddsT2','NA','source1','source2'])
DATA.index=range(len(DATA))
```

```python
DATA
```

```python
'''subset by match'''

def match_data(date, t1, t2):
    df=DATA.loc[(DATA.ix[:,0]==date) & (DATA.ix[:,4]==t1) & (DATA.ix[:,5]==t2)]
    return(df)
match_data('Apr 2', 'Kinguin', 'Encore')
```

```python
match_data('Apr 2', 'Kinguin', 'Encore').to_csv('match_example1.csv')
```

```python
'''subset by bookie''' 

def bookie(book, date, t1, t2):
    df2=match_data(date, t1, t2).loc[match_data(date, t1, t2).ix[:,2]==book]
    df2=df2.reset_index(drop=True)
    return(df2)
bookie('10Bet','Apr 2', 'Kinguin', 'Encore')
```

```python
def graph_maker(date, t1, t2):
    import plotly.plotly as py
    import plotly.graph_objs as go
    py.sign_in('polaczmaks', '5GR7uBJKEgD2IR3L7xky')

    title = 'BetterBets graph'

    labels = ['10Bet', 'Bet365', 'Betway', 'Pinnacle']

    colors = ['rgba(67,67,67,1)', 'rgba(115,115,115,1)', 'rgba(49,130,189, 1)', 'rgba(189,189,189,1)']

    mode_size = [8, 8, 12, 8]

    line_size = [2, 2, 4, 2]

    x_data = [
        bookie('10Bet',date, t1, t2).index.values.tolist(),
        bookie('Bet365',date, t1, t2).index.values.tolist(),
        bookie('Betway',date, t1, t2).index.values.tolist(),
        bookie('Pinnacle',date, t1, t2).index.values.tolist()
    ] 

    y_data = [
        bookie('10Bet',date, t1, t2)['OddsT1'].tolist(),
        bookie('Bet365',date, t1, t2)['OddsT1'].tolist(),
        bookie('Betway',date, t1, t2)['OddsT1'].tolist(),
        bookie('Pinnacle',date, t1, t2)['OddsT1'].tolist(),
    ]

    traces = []

    for i in range(len(labels)):
        traces.append(go.Scatter(
            x=x_data[i],
            y=y_data[i],
            mode='lines',
            line=dict(color=colors[i], width=line_size[i]),
            connectgaps=True,
        ))

        traces.append(go.Scatter(
            x=[x_data[i][0], x_data[i][-1]],
            y=[y_data[i][0], y_data[i][-1]],
            mode='markers',
            marker=dict(color=colors[i], size=mode_size[i])
        ))

    layout = go.Layout(        
        xaxis=dict(
            showline=True,
            showgrid=False,
            showticklabels=True,
            linecolor='rgb(204, 204, 204)',
            linewidth=2,
            autotick=False,
            ticks='outside',
            tickcolor='rgb(204, 204, 204)',
            tickwidth=2,
            ticklen=5,
            tickfont=dict(
                family='Arial',
                size=12,
                color='rgb(82, 82, 82)',
            ),
            title='Time'
        ),
        yaxis=dict(
            showgrid=False,
            zeroline=False,
            showline=False,
            showticklabels=False,
        ),
        autosize=False,
        margin=dict(
            autoexpand=False,
            l=100,
            r=20,
            t=110,
        ),
        showlegend=False,
    )

    annotations = []

    # Adding labels
    for y_trace, label, color in zip(y_data, labels, colors):
        # labeling the left_side of the plot
        annotations.append(dict(xref='paper', x=0.05, y=y_trace[0],
                                      xanchor='right', yanchor='middle',
                                      text=label.format(y_trace[0]),
                                      font=dict(family='Arial',
                                                size=16,
                                                color=colors,),
                                      showarrow=False))

    # Title

    annotations.append(dict(xref='paper', yref='paper', x=0.5, y=1.05,
                                  xanchor='middle', yanchor='bottom',
                                  text=date + ','+ ' '+ t1 + ' ' + 'vs' + ' ' + t2 + ':' + 'Odds' + ' ' +'Over' + ' ' 'Time',
                                  font=dict(family='Arial',
                                            size=30,
                                            color='rgb(37,37,37)'),
                                  showarrow=False))
    # Source
    annotations.append(dict(xref='paper', yref='paper', x=0.5, y=-0.18,
                                  xanchor='center', yanchor='top',
                                  text='BetterBets Pty Ltd',
                                  font=dict(family='Arial',
                                            size=12,
                                            color='rgb(150,150,150)'),
                                  showarrow=False))

    layout['annotations'] = annotations

    fig = go.Figure(data=traces, layout=layout)
    return(py.iplot(fig, filename='news-source'))
    
    

graph_maker('Apr 2', 'Kinguin', 'Encore')    
```
